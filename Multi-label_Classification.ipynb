```python
from sklearn.model_selection import train_test_split
X = df.drop('species', axis=1) # df.iloc[:,0:2]
y = df['species'] # df.iloc[:,-1]
X = df.drop('species', axis=1)
y = df['species']
X_train, X_holdout, y_train, y_holdout = train_test_split(X, y, test_size=0.15, random_state = 54)
len(X_train), len(y_train), len(X_holdout), len(y_holdout)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(multi_class='multinomial', solver='sag', penalty ='l2', max_iter=5000)
model.fit(X_train, y_train)
```

```python
print("Model Coefficients:\n",model.coef_)
```

```python
print("Model Intercepts:\n",model.intercept_)
```

```python
#  Calculate Accuracy score on entire dataset using cross_val_score() met
from sklearn.model_selection import cross_val_score
scores_array = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
scores_array
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score, accuracy_score

# Define the scoring metrics
scoring = {
    'precision': make_scorer(precision_score, average='micro'),
    'recall': make_scorer(recall_score, average='micro'),
    'f1': make_scorer(f1_score, average='micro'),
    'accuracy': make_scorer(accuracy_score)
}


# Perform cross-validation with the specified metrics
scores = cross_validate(model, X_train, y_train, cv=5, scoring=scoring, return_train_score=True)

# Access the results
precision_mean = scores['test_precision'].mean()
recall_mean = scores['test_recall'].mean()
f1_mean = scores['test_f1'].mean()
accuracy_mean = scores['test_accuracy'].mean()

print(f"Mean Precision: {precision_mean}")
print(f"Mean Recall: {recall_mean}")
print(f"Mean F1 Score: {f1_mean}")
print(f"Mean Accuracy: {accuracy_mean}")
```

```python
y_pred = model.predict(X_holdout)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
macro_precision = precision_score(y_holdout, y_pred, average='macro')
macro_recall = recall_score(y_holdout, y_pred, average='macro')
macro_f1 = f1_score(y_holdout, y_pred, average='macro')
print("Macro Precision score: ", macro_precision)
print("Macro Recall score: ", macro_recall)
print("Macro F1 score: ", macro_f1)
```
